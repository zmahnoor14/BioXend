```python
from rdkit import Chem
import pandas as pd
from rdkit.Chem import PandasTools
import os
from rdkit import RDConfig
import itertools
```

```python
sdfFile = os.path.join(RDConfig.RDDataDir,'/g/zimmermann/Members/zulfiqar/ZM000_PistoiaPrj/GutMicrobiome/myCompounds.sdf')
```

```python
frame = PandasTools.LoadSDF(sdfFile, includeFingerprints=True)
```

```python
frame = frame[~frame.ID.str.contains("TROSPIUM_CHLORIDE")]
frame = frame[~frame.ID.str.contains("QUINACRINE")]
frame = frame[~frame.ID.str.contains("DIGITOXIN")]
frame = frame[~frame.ID.str.contains("DIGOXIN")]
```

```python
ID = ["BETAMETHASONE", "DIHYDROERGOTAMINE_MESYLATE", "MESORIDAZINE_BESYLATE", "ARIPIPRAZOLE", "PROCHLORPERAZINE_EDISYLATE"]
SMILES = ["CC1CC2C3CCC4=CC(=O)C=CC4(C3(C(CC2(C1(C(=O)CO)O)C)O)F)C",
"CC1(C(=O)N2C(C(=O)N3CCCC3C2(O1)O)CC4=CC=CC=C4)NC(=O)C5CC6C(CC7=CNC8=CC=CC6=C78)N(C5)C.CS(=O)(=O)O",
"CN1CCCCC1CCN2C3=CC=CC=C3SC4=C2C=C(C=C4)S(=O)C.C1=CC=C(C=C1)S(=O)(=O)O",
"C1CC(=O)NC2=C1C=CC(=C2)OCCCCN3CCN(CC3)C4=C(C(=CC=C4)Cl)Cl",
"CN1CCN(CC1)CCCN2C3=CC=CC=C3SC4=C2C=C(C=C4)Cl.C(CS(=O)(=O)O)S(=O)(=O)O"]
```

```python
ROMol = [Chem.MolFromSmiles(i) for i in SMILES]
```

```python
salt_names = ["CALCIUM", "PHOSPHATE", "SULFATE", "MALEATE", "FUMARATE", "MESYLATE", "CITRATE", "HYDROBROMIDE", "BROMIDE", "SUCCINATE", "TARTRATE", "TROMETHAMINE", 
"OXALATE", "CHLORIDE", "ERBUMINE", "SODIUM", "METHANESULFONATE", "BENZOATE", "HYDRCHLORIDE", "BESYLATE", "EDISYLATE"]
```

```python
for i, row in frame.iterrows():
    found_elements = [element for element in salt_names if element in frame["ID_pubchem"][i]]
    stringuse = str(frame["ID_pubchem"][i])
    if len(found_elements)>1:
        found_elements = max(found_elements, key=len)
        frame["ID_pubchem"][i] = stringuse.replace(found_elements, "").strip()
    elif len(found_elements) == 1: 
        frame["ID_pubchem"][i] = stringuse.replace(found_elements[0], "").strip()
```
